```python
def upload_files(filename):
    """
    Upload file to colab runtime
    """
    print(f"uploading....{filename}....")
    files.upload()

    if not os.path.isfile(filename):
        print(f"{filename} does not exist. Please upload mannually...")
        exit()
```

```python
upload_files("stock_data.csv")
```
