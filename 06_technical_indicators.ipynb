### Moving Average

* https://www.investopedia.com/terms/m/movingaverage.asp

M Ani = Ci-(n-1) + Ci-(n-2) + ... + Ci-1 + Ci / n

* M Ani: Moving average of period n at trading day i
* n: Period
* Ci: Closing price at trading day i


M A10i = Ci-9 + Ci-8 + ... + Ci-1 +Ci / n


### Signal

* Close price > MAn

### MACD (Moving Average Convergence / Divergence)

* MACD is a lagging indicator when trading on the crossovers
* https://www.investopedia.com/terms/m/macd.asp

#### Calculation
(12-26-9 MACD (default), 8-17-9 MACD (more responsive))

* MACD = 12-Period EMA - 26-Period EMA
* Signal line 9-Period EMA of MACD

#### Signal

* MACD > Signal line

```python
fb = pd.read_csv("ticker-fb.csv", index_col=0, parse_dates=True)
fb.head()
```

```python
exp1 = fb['Close'].ewm(span=8, adjust=False).mean()
exp2 = fb['Close'].ewm(span=17, adjust=False).mean()
fb['MACD'] = exp1 - exp2
fb['Signal line'] = fb['MACD'].ewm(span=9, adjust=False).mean()
```

```python
fb.tail()
```

```python
fig, ax = plt.subplots()
fb.loc['2020-01-01':][['MACD', 'Signal line']].plot(ax=ax)
fb.loc['2020-01-01':]['Close'].plot(ax=ax, secondary_y=True, alpha=0.5);
```

```python
fb['MACD'] > fb['Signal line']
```

### Stochastic oscillator

* Lagging indicator
* https://www.investopedia.com/terms/s/stochasticoscillator.asp

#### Calculations
* 14-high: Maximum of last 14 trading days
* 14-low: Minimum of last 14 trading days
* %K: (Last close - 14-low) * 100 / (14-high - 14-low)
* %D: Simple Moving Average of %K

#### Signal
* %K > %D

```python
ticker = pd.read_csv("ticker-fb.csv", index_col=0, parse_dates=True)
ticker.head()
```

```python
high14 = ticker['High'].rolling(14).max()
low14 = ticker['Low'].rolling(14).min()
# %K: (Last close - 14-low) * 100 / (14-high - 14-low)
pct_k = (ticker['Close'] - low14) * 100 / (high14 - low14)
pct_d = pct_k.rolling(3).mean()
```

```python
ticker['%K'] = pct_k
ticker['%D'] = pct_d
ticker['Signal'] = pct_k > pct_d
```

```python
ticker.tail()
```

```python
fig, ax = plt.subplots()
ticker.loc['2020-01-01':][['%K', '%D']].plot(ax=ax, alpha=0.5)
ax.axhline(80, c='r', alpha=0.3)
ax.axhline(20, c='r', alpha=0.3)
ticker.loc['2020-01-01':]['Close'].plot(ax=ax, secondary_y=True);
```

```python
data = pd.read_csv("ticker-fb.csv", index_col=0, parse_dates=True)
data.head()
```

```python
delta = data['Close'].diff()
```

```python
delta.head()
```

```python
up = delta.clip(lower=0)
down = -1 * delta.clip(upper=0)
```

```python
up.head(), down.head()
```

```python
ema_up = up.ewm(com=13, adjust=False).mean()
ema_down = down.ewm(com=13, adjust=False).mean()
```

```python
ema_up.head(), ema_down.head()
```

```python
rs = ema_up / ema_down
```

```python
rs.tail()
```

```python
data['RSI'] = 100 - (100 / (1 + rs))
```

```python
data.tail()
```

```python
fig, ax = plt.subplots()
data['RSI'].plot(ax=ax);
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
data['Close'].plot(ax=ax[0], title="Close")
data["RSI"].plot(ax=ax[1], title="RSI")
ax[1].axhline(70, c='r', alpha=0.3, linestyle='--')
ax[1].axhline(30, c='r', alpha=0.3, linestyle='--')
plt.tight_layout();
```
