### Matplotlib - Part II

```python
data = pd.read_csv("result-full.csv", index_col=0, header=[0, 1])
data.head()
```

```python
data.iloc[0][3:]
```

```python
def plot(df, start=0, nrows=3, ncols=3):
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols)
    for i in range(nrows):
        for j in range(ncols):
            pos = start + j + i * ncols
            frame = df.iloc[pos][3:].copy()
            name = frame.name
            frame = frame.unstack(1)
            frame.plot.bar(ax=ax[i, j])
            ax[i, j].set_title(name)
            ax[i, j].axhline(0.1, color='r', alpha=0.3)
    plt.rcParams.update({'font.size': 3})
    plt.tight_layout()
```

```python
plot(data)
```

```python
plot(data, start=9)
```

```python
plot(data.sort_values(('BVPS', '1Y CAGR'), ascending=False))
```

```python
plot(data.sort_values(('BVPS', '10Y CAGR'), ascending=False))
```

```python
writer = pd.ExcelWriter("data2.xlsx", engine="xlsxwriter")
data.to_excel(writer, sheet_name="Main")

workbook = writer.book
worksheet = writer.sheets["Main"]

# Create a format for a green cell
green_cell = workbook.add_format({
    'bg_color': '#C6EFCE',
    'font_color': '#006100'
})

# create a format for a red cell
red_cell = workbook.add_format({
    'bg_color': '#FFC7CE',
    'font_color': '#9C0006'
})

# create a conditional formatted of type formula
worksheet.conditional_format('B4:D53', {
    'type': 'cell',
    'criteria': '=',
    'value': True,
    'format': green_cell
})

# create a conditional formatted of type formula
worksheet.conditional_format('B4:D53', {
    'type': 'cell',
    'criteria': '=',
    'value': False,
    'format': red_cell
})

# create a conditional formatted of type formula
worksheet.conditional_format('E4:P53', {
    'type': 'cell',
    'criteria': '>=',
    'value': 0.1,
    'format': green_cell
})

# create a conditional formatted of type formula
worksheet.conditional_format('E4:P53', {
    'type': 'cell',
    'criteria': '<',
    'value': 0.1,
    'format': red_cell
})

writer.close()
```

```python
def read_annual_data(csv_file, ticker):
    upload_files(csv_file)
    data = pd.read_csv(csv_file, index_col=0)
    data = data[ticker]
    data = data.pct_change()
    data = data.loc[2009:2019]
    return data
```

```python
def read_quartely_data(csv_file, ticker):
    upload_files(csv_file)
    data = pd.read_csv(csv_file, index_col=0, parse_dates=True)
    data = data[ticker]
    data = data.resample('Y', convention='end').agg('mean')
    data = data.pct_change()
    data = data.loc['2009-01-01': '2019-12-31']
    data.index = data.index.year
    return data
```

```python
def read_quartely_ratio(csv_file, ticker):
    upload_files(csv_file)
    data = pd.read_csv(csv_file, index_col=0, parse_dates=True)
    data = data[ticker]
    data = data.resample('Y', convention='end').agg('mean')
    data = data.loc['2009-01-01': '2019-12-31']
    data.index = data.index.year
    return data
```

```python
ticker = "NFLX"
```

```python
rev = read_annual_data("50-rev.csv", ticker)
eps = read_annual_data("50-eps.csv", ticker)
fcf = read_annual_data("50-fcf.csv", ticker)
```

```python
bvps = read_quartely_data("50-bvps.csv", ticker)
```

```python
roi = read_quartely_ratio("50-roi.csv", ticker)
de = read_quartely_ratio("50-de-ratio.csv", ticker)
pe = read_quartely_ratio("50-pe-ratio.csv", ticker)
current = read_quartely_ratio("50-current-ratio.csv", ticker)
```

```python
def add_annual_data(writer, df, sheet_name):
    df.to_excel(writer, sheet_name=sheet_name)

    workbook = writer.book
    worksheet = writer.sheets[sheet_name]

    # create a new chart object
    chart = workbook.add_chart({'type': 'line'})

    # Add a series to the chart
    chart.add_series({
        'name': sheet_name,
        'categories': [sheet_name, 1, 0, 11, 0],
        'values': [sheet_name, 1, 1, 11, 1]
    })

    # to set the labels on x axis not on 0 and rotate
    chart.set_x_axis({
        'label_position': 'low',
        'num_font': {'rotation': 45}
    })

    # insert the chart into the worksheet.
    worksheet.insert_chart('D1', chart)
```

```python
writer = pd.ExcelWriter("data3.xlsx", engine="xlsxwriter")
add_annual_data(writer, rev, "Revenue")
add_annual_data(writer, eps, "EPS")
add_annual_data(writer, fcf, "FCF")
add_annual_data(writer, bvps, "BVPS")
add_annual_data(writer, roi, "ROI")
add_annual_data(writer, de, "DE")
add_annual_data(writer, pe, "PE")
add_annual_data(writer, current, "Current")
writer.close()
```
