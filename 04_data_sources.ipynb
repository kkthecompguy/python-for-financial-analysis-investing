### Web Scrapping

* Be ethical
* Do not use for commercial purposes
* Use only for private

```python
url = "https://en.wikipedia.org/wiki/Wikipedia:Fundraising_statistics"

tables = pd.read_html(url)
```

```python
len(tables)
```

```python
fundraising = tables[0]
```

```python
fundraising.head()
```

```python
fundraising.dtypes
```

```python
fundraising['Rev'] = fundraising['Revenue'].str[2:]
```

```python
fundraising.head()
```

```python
fundraising['Rev'] = fundraising['Rev'].str.replace(',', '')
```

```python
fundraising.head()
```

```python
fundraising.dtypes
```

```python
fundraising['Rev'] = fundraising['Rev'].apply(pd.to_numeric)
```

```python
fundraising.dtypes
```

```python
fundraising['Exp'] = fundraising['Expenses'].str[2:]
```

```python
fundraising.head()
```

```python
fundraising['Exp'] = fundraising['Exp'].str.replace(',', '')
```

```python
fundraising.loc[0, 'Exp'] = 'spam'
```

```python
fundraising.head()
```

```python
fundraising['Exp'] = pd.to_numeric(fundraising['Exp'], errors='coerce')
```

```python
fundraising.head()
```

```python
tickers = ['NVDA', 'HP', 'META', 'GOOG']
data = pdr.get_data_yahoo(tickers, start, end)
```

```python
data.head()
```

```python
data["Close"]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
data["Close"]["NVDA"].plot(ax=ax[0,0], title="NVDA")
data["Close"]["GOOG"].plot(ax=ax[0,1], title="GOOG")
data["Close"]["HP"].plot(ax=ax[1,0], title="HP")
data["Close"]["META"].plot(ax=ax[1,1], title="META")
plt.tight_layout()
```

```python
nvda = data.xs("NVDA", axis=1, level=1)
goog = data.xs("GOOG", axis=1, level=1)
meta = data.xs("META", axis=1, level=1)
hp = data.xs("HP", axis=1, level=1)
```

```python
nvda.to_csv("nvda.csv")
goog.to_csv("goog.csv")
meta.to_csv("meta.csv")
hp.to_csv("hp.csv")
```
