### Calculate a Fair Price (Intrinsic Value)

Consider the following 4 key metrics to calculate a fair price

 * Current EPS ----------------> growth rate
 * Future EPS ------------------> 10yrs ahead of time
 * Future P/E  -----------------> Future EPS * Future PE
 * Price <--------------------- 10yrs back

```python
# upload 50-pe-ratio.csv
files.upload()

check_file_exist("50-pe-ratio.csv")
```

```python
pe = pd.read_csv("50-pe-ratio.csv", index_col=0, parse_dates=True)
pe.head()
```

```python
pe.sort_index(ascending=False, inplace=True)
```

```python
pe.head()
```

```python
data = pe.resample('Q', convention='end').agg('mean')
```

```python
data = data.loc['2010-01-01':]
```

```python
data.head()
```

```python
data = data.transpose()
```

```python
data['Mean'] = data.mean(axis=1)
```

```python
data.head()
```

```python
data['Mean'].to_csv("result-pe-avg.csv")
```

```python
eps = pd.read_csv("50-eps.csv", index_col=0)
```

```python
current_eps = eps.transpose()[2019]
```

```python
current_eps = current_eps[current_eps > 0].copy()
```

```python
bvps = pd.read_csv("result-bvps.csv", index_col=0)
```

```python
eps_gr = bvps.min(axis=1)
```

```python
eps_gr = eps_gr[eps_gr > 0].copy()
```

```python
future_eps = current_eps * (1 + eps_gr) ** 10
```

```python
pe = pd.read_csv("result-pe-avg.csv", index_col=0)
```

```python
future_pe = pd.concat([pe, eps_gr*100*2], axis=1)
```

```python
future_pe = future_pe.min(axis=1)
```

```python
future_price = future_eps * future_pe
```

```python
price = future_price * (1/1.15)**10
```

```python
price.to_csv("result-price.csv")
```
